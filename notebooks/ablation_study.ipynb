# Ablation Study Analysis

We analyse the ablation results saved to results/metrics/ablation_results.csv.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

abl = pd.read_csv('../results/metrics/ablation_results.csv')
print(abl[['ablation', 'description', 'mean_r2', 'delta_r2']].to_string(index=False))
```

## Ablation Bar Chart

```python
df_abl = abl[abl['ablation'] != 'full_pipeline'].copy()
fig, ax = plt.subplots(figsize=(9, 5))
colors = ['#d62728' if x < 0 else '#2ca02c' for x in df_abl['delta_r2']]
bars = ax.bar(df_abl['ablation'], df_abl['delta_r2'], color=colors, edgecolor='white')
ax.axhline(0, color='black', lw=0.8)
ax.set_ylabel('Delta R^2 vs full pipeline')
ax.set_title('Ablation Study: Impact of Removing Each SP Component')
for bar, val in zip(bars, df_abl['delta_r2']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + (0.001 if val >= 0 else -0.01),
            f'{val:+.4f}', ha='center', fontsize=10, fontweight='bold')
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Key Findings

The ablation study reveals that **Poisson event modelling is the single most critical SP component**.
Removing it collapses R^2 from 0.795 to 0.224 (delta = -0.571). Without discrete disturbance events,
the features light_n_events and noise_n_events become zero for all sessions, removing the primary
drivers of sleep quality differentiation.

The Butterworth LPF provides a modest contribution (delta = -0.007), confirming that temporal
smoothness improves feature quality. Pink noise and HVAC sawtooth have negligible predictive impact
but contribute to physical realism.
